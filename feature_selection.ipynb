```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
```

```python
tickers = ['XOM', 'CVS']
```

```python
xom_data = pd.read_csv('data/others/XOM.csv')
cvs_data = pd.read_csv('data/others/CVS.csv')

weather = pd.read_csv('data/Weather.csv')
```

```python
xom_data = xom_data.rename(columns={'Unnamed: 0':'Date'}).set_index('Date').dropna()
cvs_data = cvs_data.set_index('Date').dropna()

weather = weather.rename(columns={'begin_date':'Date'}).set_index('Date').dropna()
```

```python
xom_data = xom_data.join(weather)
cvs_data = cvs_data.join(weather)
```

```python
xom_data.to_csv('data/XOM.csv')
cvs_data.to_csv('data/CVS.csv')
```

```python
xom_columns = ['XOM.Open', 'XOM.High', 'XOM.Low', 'XOM.Close', 'XOM.Volume',
               'XOM.fans_fb', 'XOM.engagement_score_fb',
               'XOM.followers_count_twtt', 'XOM.engagement_score_twtt',
               'MAGNITUDE', 'Flood', 'Heavy Snow', 'High Wind', 'Winter Weather', 'Lightning',
               'Blizzard']

cvs_columns = ['CVS.Open', 'CVS.High', 'CVS.Low', 'CVS.Close', 'CVS.Volume',
               'CVS.fans_fb', 'CVS.engagement_score_fb',
               'CVS.followers_count_twtt', 'CVS.engagement_score_twtt',
               'MAGNITUDE', 'Flood', 'Heavy Snow', 'High Wind', 'Winter Weather', 'Lightning',
               'Blizzard']
```

```python
xom_data[xom_columns].to_csv('data/XOM.csv')
cvs_data[cvs_columns].to_csv('data/CVS.csv')
```
